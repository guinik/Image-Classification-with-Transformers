```python
class dataAugmentation(tf.keras.Sequential):
    def __init__ (self,image_size,height_factor=0.1,width_factor=0.1,factor=0.02,flip='horizontal'):
        super().__init__()
        #WE DO SOME AUGMENTATION SO WE TRAIN THE MODEL TO MORE RANDOM SITUATIONS
        self.Normalization=layers.Normalization()
        self.Resizing=layers.Resizing(image_size,image_size)
        self.RandomFlip=layers.RandomFlip(flip)
        self.RandomRotation=layers.RandomRotation(factor)
        self.RandomZoom=layers.RandomZoom(height_factor,width_factor)
        
    def __call__(self,data):
        result=[]
        #WE DO FOR EVERY IMAGE THE AUGMENTATION AND SAVE IT
        for image in data:
            self.Normalization.adapt(image)
            image=self.Normalization(image)#this way out std and mean are around 1,0
            image=self.Resizing(image)
            image=self.RandomFlip(image)
            image=self.RandomRotation(image)
            image=self.RandomZoom(image)
            result.append(image)
        return tf.convert_to_tensor(result)
```

```python
image_size=128 #we will do 128X128
test=dataAugmentation(image_size)#TEST is our converter
```

Now we need to create patches, dividing the images. They can be thought, if one is more familiar with translation transformers, as seperating the images into words, and then we encode those words.
We will use tf.image.extract_patches, it does exactly what we need.

We will divide it in 16x16 subsquares. And then we will encode it 

```python
class Patches:
    def __init__(self,patch_size):
        self.patch_size=patch_size
        
    def __call__(self,data):
        
        batch_size=tf.shape(data)[0]
        #this function will flatten a  patch_size x patch_size square in stride steps in the last dimension
        patches=tf.image.extract_patches(images=data,
                                        sizes=[1,self.patch_size,self.patch_size,1],
                                        strides=[1,self.patch_size,self.patch_size,1],
                                        rates=[1,1,1,1],
                                        padding='VALID')
        
        #This will turn the (batch, 128,128,3) into (batch,8,8,3*16*16) but to feed our network we will want it as
        #(batch,numpatches=8*8,patchdim=3*16*16) , we assumend 16 patch size
        patchdim=tf.shape(patches)[-1]
        patches=tf.reshape(patches,[batch_size,-1,patchdim])
        return patches
        
```

```python
Patcher=Patches(16)#for our specific purposes it will be 16 -> 16x16 squares
#PATCHER(TRAIN) will be our training data!
```

We will use this simple embedding. Where we use the built-in embedding then sum the lineally projected and the position embedding. This is a learnable position_embedding

```python
class PatchEncoder(layers.Layer):
    def __init__(self,num_patches,projection_dim):
        super().__init__()
        
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )
    
    def call(self, patch):
        #simply que apply the embedding to get our d_model dimension and the poisitional embedding is a simple one, 
        #for every subsquare a number , and we use embedding from keras
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded
        
        
```

We will also introduce the MLP layers, they are simply a Dense and a droupout. And we will use tanh as activation

```python
class MLP(layers.Layer):
    def __init__(self, dropout_rate,modeldim):
        super().__init__()
        self.Dense=layers.Dense(modeldim)
        self.dropout1=layers.Dropout(dropout_rate)
        
        
    def __call__(self, x ,training):
        x=self.Dense(x)#simple perceptron 
        x=self.dropout1(x,training)
        return x
```

```python
class MultiAttention(tf.keras.layers.Layer):
    #This can be found more in detail in my other repository Transformer for translation
    def __init__(self,modeldim,layersnum): 
        # we need to inherit all of the keras information of th super
        super().__init__()
        #now we can start working
        
        #WEIGHTS FOR Q,K,V
        self.wq=tf.keras.layers.Dense(modeldim)
        self.wk=tf.keras.layers.Dense(modeldim)
        self.wv=tf.keras.layers.Dense(modeldim)
                          
            
        #WEIGHTS FO AFTER CONCATENTION
        self.dense=tf.keras.layers.Dense(modeldim)
        
        #extra information
        
        self.modeldim=modeldim
        self.layersnum=layersnum
        assert modeldim % self.layersnum == 0
        self.depth=modeldim//self.layersnum
        
        
    #WE NEED TO SPLIT INTO modeldim /layresnum
    
    #All this part is heavily optimized thanks to the tensorflow tutorial for transformers
    #The main idea is to split the last dimension into (number of layers, model dimension)
    def split_heads(self, x, batch_size):
        #Split the last dimension into (layersnum, depth).
        #Transpose the result such that the shape is (batch_size, num_heads, seq_len, depth)
        #
        x = tf.reshape(x, (batch_size, -1, self.layersnum, self.depth))
        return tf.transpose(x, perm=[0, 2, 1, 3])

    def call(self, v, k, q, mask):
        batch_size = tf.shape(q)[0]

        q = self.wq(q)  # (batch_size, seq_len, d_model)
        k = self.wk(k)  # (batch_size, seq_len, d_model)
        v = self.wv(v)  # (batch_size, seq_len, d_model)

        q = self.split_heads(q, batch_size)  # (batch_size, layersnum, seq_len_q, depth)
        k = self.split_heads(k, batch_size)  # (batch_size, layersnum, seq_len_k, depth)
        v = self.split_heads(v, batch_size)  # (batch_size, layersnum, seq_len_v, depth)

        # scaled_attention.shape == (batch_size, layersnum, seq_len_q, depth)
        # attention_weights.shape == (batch_size, layersnum, seq_len_q, seq_len_k)
        scaled_attention, attention_weights = dotproductattention(
            q, k, v, mask)

        scaled_attention = tf.transpose(scaled_attention, perm=[0, 2, 1, 3])  # (batch_size, seq_len_q, num_heads, depth)

        concat_attention = tf.reshape(scaled_attention,
                                      (batch_size, -1, self.modeldim))  # (batch_size, seq_len_q, d_model)

        output = self.dense(concat_attention)  # (batch_size, seq_len_q, d_model)

        return output, attention_weights
def dotproductattention(q,k,v,mask):
    
    
    #First we need to matmul q,k transposing k
    resmatmul=tf.matmul(q,k,transpose_b=True)
    
    #then we need the dimension of k d_k 
    kdim=float(tf.shape(k)[-1])
    resmatmul=resmatmul/(tf.math.sqrt(kdim))
    
    #It may be useful to include the situation where mask is None
    if mask is not None:
        resmatmul += mask*-1e9
    
    ressoftmax=tf.nn.softmax(resmatmul,axis=-1) # we need to do softmax on the last dimension
    return tf.matmul(ressoftmax,v), ressoftmax
    
    
```

```python
class EncoderLayer(layers.Layer):
    def __init__(self, d_model, num_heads, rate=0.1):
        super(EncoderLayer, self).__init__()
        self.layernorm1 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        
        self.mha = MultiAttention(d_model, num_heads)
        self.dropout1 = tf.keras.layers.Dropout(rate)
        self.mlp=MLP(rate,d_model)
        
    def call(self, x, training, mask): #Mask will be None in this case
        #WE FOLLOW THE STRUCTURE GIVEN IN THE PAPER
        
        out=self.layernorm1(x)
        attn_output, _ = self.mha(out, out, out, mask)  # (batch_size, patchnum, d_model) we do an attention
        attn_output = self.dropout1(attn_output, training=training)#droupout for better training
        out1 =(x + attn_output)  # (batch_size,  patchnum, d_model) normalization
         
        norm2 = self.layernorm2(out1)  # (batch_size,  patchnum, d_model) 

        ffn_output = self.mlp(norm2,training)  # (batch_size, patchnum , d_model) feedforwards 
        outfinal=out1+ffn_output
        return outfinal
```

We finally build our model puttin everything together. We will at the end flatten everything and apply a MLP that will give us the 120 dimension that we need for Categorical Classification

```python
class Model(tf.keras.Model):
    
    def __init__(self,num_patches, num_layers, d_model, num_heads,num_classes, rate=0.1):
        super(Model, self).__init__()

        self.d_model = d_model
        self.num_layers = num_layers

        self.enc_layers = [EncoderLayer(d_model, num_heads, rate)
                           for _ in range(num_layers)]

        self.patch_encoder=PatchEncoder(num_patches,d_model)
        self.softmax=layers.Softmax()

        self.flatten=layers.Flatten()
        
        self.headmlp=MLP(rate,num_classes)
        
    def call(self, x, training, mask=None):
        encoded=self.patch_encoder(x)#we encode it
        result=encoded
        #iterate for every layer of encoders
        for i in range(self.num_layers):
            result = self.enc_layers[i](result, training, mask=None)
        
        result=self.flatten(result)#we flatten
        
        result=self.headmlp(result,training)#we get the desired dimensions
        return result #we are done!
    
```

Our parameters are going to be 5 layers, 64 is our number of patches 8x8, and d_model is going to be 128, num _heads is gonna be 4, and num_classes is 120 types of dogs.

```python
Modelo=Model(64,5,128,4,120)# we build the model
```

All thats left is train it!


```python
learning_rate = 0.00000001#pretty sensible to learning rate
def run_ex(model):
    
    optimizer = tf.keras.optimizers.Adam()

    model.compile(
        optimizer=optimizer,
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
        metrics=[
            tf.keras.metrics.SparseCategoricalAccuracy(name="accuracy"),
            tf.keras.metrics.SparseTopKCategoricalAccuracy(5, name="top-5-accuracy"),
        ],
    )
    checkpoint_path = "./checkpoints/train"

    ckpt = tf.train.Checkpoint(model=model,
                           optimizer=optimizer)

    ckpt_manager = tf.train.CheckpointManager(ckpt, checkpoint_path, max_to_keep=5)

    if ckpt_manager.latest_checkpoint:
        ckpt.restore(ckpt_manager.latest_checkpoint)
        print('Latest checkpoint restored!!')
    
    model.fit(x=Patcher(train),y=labels,epochs=100,batch_size=25,steps_per_epoch=1)
    ckpt_save_path = ckpt_manager.save()
```

```python
#this will train it 
run_ex(Modelo)
```
